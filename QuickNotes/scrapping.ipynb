```python
import requests
```

```python
import bs4
```

```python
import lxml
```

```python
result = requests.get("https://es.wikipedia.org/wiki/League_of_Legends")
```

```python
type(result)
```

```python
result.text
```

```python
# Now we have the string that we want, we need to convert it to a 'soup' with bs4

soup = bs4.BeautifulSoup(result.text,'lxml')
```

```python
soup
```

With that method, we are able to transform the string into a model that is readable, as the source code of the web page. Now we are going to grab things from the html doc

```python
soup.select('title') #We need to pass in the 'tag' we are interested for

#It passes out a list, since it could be more tags in the page
```

If we need to return only the string, or assign it to a variable:

```python
soup.select('title')[0].getText()
```

```python
souptitle = soup.select('title')[0].getText()
```

```python
souptitle
```

With this method, we can get the strings we need or pass to the code

It is usefull to check the html syntax table for soup elements

```python
soup.select('.toctext')

# Here we are grabbing the index, we can see in the webpage, that the tag is 'toctext'
```

```python
soupindex = soup.select('.toctext')[1].getText()

soupindex
```

```python
# we can print every index title by a simple for loop

for item in soup.select('.toctext'):
    print(item.text)
```

```python
soup.select('img')[0]

```

as we see, the img has a lot more of info that we need for grabbing an image, we are going to look for "src" since is the source of the url related image

```python
# https://upload.wikimedia.org/wikipedia/commons/thumb/d/dc/Map_of_MOBA.svg/220px-Map_of_MOBA.svg.png 
# this is the image we want, but in the html we hace thumbimage as the tag related

soup.select('.thumbimage')
```

NOTE: Important, when we want to tag somthing in a class, we need to put a dot '.' before the tag

Here we are seeing that the one we are looking for, is the first one


```python
mapa = soup.select('.thumbimage')[0]

mapa
```

```python
# Since this is a tag bs4 element, we can make diccionary calls to search what we are looking for

mapa['src']
```
